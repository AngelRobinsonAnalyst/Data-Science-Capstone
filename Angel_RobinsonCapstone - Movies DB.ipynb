## Task 8: 
- Identify and display the names of all production companies along with the number of times they appear in the dataset.

```python
production_companies
```

```python
production_companies.value_counts()


# extract the production companies column as a list of dictionaries
companies = movies_df["production_companies"].apply(eval)

# create an empty dictionary to store the company names and counts
company_counts = {}

# loop through each row of the companies column
for row in companies:
    # loop through each dictionary in the row
    for company in row:
        # get the company name
        name = company["name"]
        # increment the count of the company by 1
        company_counts[name] = company_counts.get(name, 0) + 1

# create a pandas series from the company counts dictionary
company_series = pd.Series(company_counts)

# sort the series in descending order
company_series = company_series.sort_values(ascending=False)

# display the companies and counts
print(company_series)
```

## Task 10: 
> Sort the data in descending order based on revenue and filter the top 500 movies.

> Find the measures of central tendency for the following columns using the filtered data:
- Budget
- Revenue
- Runtime

```python
import statistics as st
```

```python
top_500 = movies_df.loc[:,['budget','revenue','runtime']].sort_values('revenue',ascending = False).head(500)
```

```python
# CREATE 3 EMPTY LISTS TO STORE CENTRAL TENDENCIES (measures)
mean = []
median = []
mode = []
```

```python
# ITERATE OVER EACH COL USING FOR LOOP
for col in top_500:
    mean.append(top_500.loc[:,col].mean())# calcualtes each column mean and stores it in mean list
    median.append(top_500.loc[:,col].median())# calcualtes each column median and stores it in median list
    mode.append(st.mode(top_500[col]))# calcualtes each column mode and stores it in mode list
```

```python
# CREATE DF
row_head = ['Mean','Median','Mode']
col_names = ['Budget','Revenue','Runtime']
top_500_df = pd.DataFrame([mean,median,mode],columns = col_names)
# INSERTING A NEW COLUMN INTO THE DF TO DISPLAY CENTRAL TENDENCIES
top_500_df.insert(0,'Measures',row_head,True)
```

```python
print(top_500_df)
```
